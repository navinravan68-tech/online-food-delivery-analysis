Final Amount = Order Value - Discount Applied

final varaible contains the condition isNan (isna), not NaN(notna) then applying the formula

### Handling Missing Financial Records

Financial columns (`order_value`, `discount_applied`, and `final_amount`) contained a large number of missing values.

Business validation showed that:

- `0` in `discount_applied` explicitly represents "no discount".
- `NaN` in `discount_applied` represents missing information rather than zero.
- Missing `order_value` and `final_amount` could not be reliably reconstructed.

To ensure accurate revenue, profit, and order value analysis, records with missing essential financial information were excluded from the analytical dataset.

Final cleaned dataset shape:
- Rows: 55,514
- Columns: 23

Outliers Analysis

```python
numeric_cols = [
    "order_value",
    "final_amount",
    "delivery_time_min",
    "distance_km",
    "profit_margin"
]
```

```python
pip install seaborn
```

```python
import seaborn as sns
```

```python
for col in numeric_cols:
    plt.figure(figsize=(8, 2.5))
    sns.boxplot(x=df[col])
    plt.title(f"Box Plot of {col.replace('_', ' ').title()}")
    plt.show()
```

```python
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower) | (df[col] > upper)]

    print(f"{col}")
    print(f"Lower Bound : {lower:.2f}")
    print(f"Upper Bound : {upper:.2f}")
    print(f"Outliers    : {len(outliers)}")
    print("-" * 40)
```
